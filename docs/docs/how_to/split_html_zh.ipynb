### 选择合适的分割器
- **使用 `HTMLHeaderTextSplitter` 的场景**：当你需要根据HTML文档的标题层级进行分割，并保留标题相关的元数据时。- **使用 `HTMLSectionSplitter` 的情况**：当你需要将文档分割成更大、更通用的部分时，可能是基于自定义标签或字体大小。- **使用 `HTMLSemanticPreservingSplitter` 的场景**：当你需要将文档分割成块的同时保留表格和列表等语义元素，确保它们不会被分割且上下文得以保持。

## 使用HTMLHeaderTextSplitter
[HTMLHeaderTextSplitter](https://python.langchain.com/api_reference/text_splitters/html/langchain_text_splitters.html.HTMLHeaderTextSplitter.html) 是一个"结构感知型"[文本分割器](/docs/concepts/text_splitters/)，它会在HTML元素级别分割文本，并为每个与给定文本块"相关"的标题添加元数据。该工具支持逐元素返回文本块，也可将具有相同元数据的元素合并，其目标在于：(a) 保持语义上相关联的文本（大致）成组；(b) 保留文档结构中编码的上下文丰富信息。它可以作为分块流程的一部分，与其他文本分割器配合使用。
这类似于针对Markdown文件的[MarkdownHeaderTextSplitter](/docs/how_to/markdown_header_metadata_splitter)。
要指定在哪些标题处进行分割，请在实例化 `HTMLHeaderTextSplitter` 时指定 `headers_to_split_on`，如下所示。

````python
from langchain_text_splitters import HTMLHeaderTextSplitter

headers_to_split_on = [
    ("h1", "Header 1"),
    ("h2", "Header 2"),
    ("h3", "Header 3"),
]

html_splitter = HTMLHeaderTextSplitter(headers_to_split_on)
html_header_splits = html_splitter.split_text(html_string)
html_header_splits
````

要在实例化 `HTMLHeaderTextSplitter` 时返回每个元素及其关联的标题，请指定 `return_each_element=True`：

````python
html_splitter = HTMLHeaderTextSplitter(
    headers_to_split_on,
    return_each_element=True,
)
html_header_splits_elements = html_splitter.split_text(html_string)
````

与上述情况相比，其中元素按其标题进行聚合：

````python
for element in html_header_splits[:2]:
    print(element)
````

现在每个元素都作为独立的 `Document` 返回：

````python
for element in html_header_splits_elements[:3]:
    print(element)
````

### 如何从URL或HTML文件中拆分：
要从URL直接读取内容，请将URL字符串传入`split_text_from_url`方法。
同样地，本地HTML文件也可以传递给`split_text_from_file`方法处理。

````python
url = "https://plato.stanford.edu/entries/goedel/"

headers_to_split_on = [
    ("h1", "Header 1"),
    ("h2", "Header 2"),
    ("h3", "Header 3"),
    ("h4", "Header 4"),
]

html_splitter = HTMLHeaderTextSplitter(headers_to_split_on)

# for local file use html_splitter.split_text_from_file(<path_to_file>)
html_header_splits = html_splitter.split_text_from_url(url)
````

### 如何限制分块大小：
`HTMLHeaderTextSplitter`（基于HTML标题进行分割的拆分器）可以与另一种根据字符长度限制拆分的拆分器（例如`RecursiveCharacterTextSplitter`）组合使用。
这可以通过第二个分割器的 `.split_documents` 方法实现：

````python
from langchain_text_splitters import RecursiveCharacterTextSplitter

chunk_size = 500
chunk_overlap = 30
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=chunk_size, chunk_overlap=chunk_overlap
)

# Split
splits = text_splitter.split_documents(html_header_splits)
splits[80:85]
````

### 局限性
不同HTML文档之间可能存在较大的结构差异，虽然`HTMLHeaderTextSplitter`会尝试将所有"相关"标题附加到给定文本块，但有时仍可能遗漏某些标题。例如，该算法默认信息层级中标题始终位于关联文本的"上方"节点，即前序兄弟节点、祖先节点或其组合。在以下新闻案例（截至本文撰写时）中，文档结构显示顶级标题文本虽被标记为"h1"，却位于与我们预期"统领"的文本元素*不同*的子树中——因此可以观察到"h1"元素及其关联文本未出现在分块元数据中（但在适用情况下，我们确实能看到"h2"及其关联文本）：

````python
url = "https://www.cnn.com/2023/09/25/weather/el-nino-winter-us-climate/index.html"

headers_to_split_on = [
    ("h1", "Header 1"),
    ("h2", "Header 2"),
]

html_splitter = HTMLHeaderTextSplitter(headers_to_split_on)
html_header_splits = html_splitter.split_text_from_url(url)
print(html_header_splits[1].page_content[:500])
````

## 使用 HTMLSectionSplitter
与 [HTMLHeaderTextSplitter](#using-htmlheadertextsplitter) 的概念类似，`HTMLSectionSplitter` 是一种“结构感知型”[文本分割器](/docs/concepts/text_splitters/)，它会在元素层级分割文本，并为每个与给定分块“相关”的标题添加元数据。该工具支持按章节分割 HTML 内容。
它可以逐个元素返回文本块，或将具有相同元数据的元素进行合并，其目标在于：(a) 保持语义上相关联的文本（或多或少）成组呈现；(b) 保留文档结构中所编码的富含上下文的信息。
使用 `xslt_path` 提供绝对路径来转换 HTML，以便能够根据提供的标签检测章节。默认情况下，使用 `data_connection/document_transformers` 目录中的 `converting_to_header.xslt` 文件。这是为了将 HTML 转换为更易于检测章节的格式/布局。例如，基于字体大小的 `span` 可以转换为标题标签，从而被检测为一个章节。
### 如何分割HTML字符串：

````python
from langchain_text_splitters import HTMLSectionSplitter

headers_to_split_on = [
    ("h1", "Header 1"),
    ("h2", "Header 2"),
]

html_splitter = HTMLSectionSplitter(headers_to_split_on)
html_header_splits = html_splitter.split_text(html_string)
html_header_splits
````

### 如何限制分块大小：
`HTMLSectionSplitter` 可与其他文本分割器配合使用，作为分块流程的一部分。当章节大小超过分块尺寸时，其内部会调用 `RecursiveCharacterTextSplitter` 进行处理。该工具还会根据文本的字体大小，通过预设的字体大小阈值来判断其是否构成独立章节。

````python
from langchain_text_splitters import RecursiveCharacterTextSplitter

headers_to_split_on = [
    ("h1", "Header 1"),
    ("h2", "Header 2"),
    ("h3", "Header 3"),
]

html_splitter = HTMLSectionSplitter(headers_to_split_on)

html_header_splits = html_splitter.split_text(html_string)

chunk_size = 50
chunk_overlap = 5
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=chunk_size, chunk_overlap=chunk_overlap
)

# Split
splits = text_splitter.split_documents(html_header_splits)
splits
````

## 使用 HTML语义保留分割器
`HTMLSemanticPreservingSplitter` 旨在将 HTML 内容分割成可管理的块，同时保留表格、列表和其他 HTML 组件等重要元素的语义结构。这确保了这些元素不会被分割到不同的块中，从而避免丢失上下文相关性，例如表头、列表标题等。
该分块器的核心设计目标是生成上下文相关的文本块。若使用`HTMLHeaderTextSplitter`进行常规递归分割，可能导致表格、列表等结构化元素在中间被截断，从而丢失重要上下文并产生低质量的分块结果。
`HTMLSemanticPreservingSplitter` 对于分割包含表格和列表等结构化元素的 HTML 内容至关重要，尤其是在需要保持这些元素完整的情况下。此外，它能够为特定的 HTML 标签定义自定义处理程序，使其成为处理复杂 HTML 文档的多功能工具。
**重要提示**：`max_chunk_size` 并非分块的绝对最大值限制。最大值的计算发生在保留内容不属于当前分块时，以确保这些内容不会被分割。当我们将保留数据重新加入分块时，分块大小可能会超过 `max_chunk_size`。这一点对于保持原始文档结构至关重要。

### 保留表格与列表在本示例中，我们将展示`HTMLSemanticPreservingSplitter`如何保留HTML文档中的表格和大型列表。区块大小将设置为50个字符，以此说明拆分器如何确保这些元素即使超过定义的最大区块大小也不会被分割。

````python
from langchain_text_splitters import HTMLSemanticPreservingSplitter

html_string = """
<!DOCTYPE html>
<html>
    <body>
        <div>
            <h1>Section 1</h1>
            <p>This section contains an important table and list that should not be split across chunks.</p>
            <table>
                <tr>
                    <th>Item</th>
                    <th>Quantity</th>
                    <th>Price</th>
                </tr>
                <tr>
                    <td>Apples</td>
                    <td>10</td>
                    <td>$1.00</td>
                </tr>
                <tr>
                    <td>Oranges</td>
                    <td>5</td>
                    <td>$0.50</td>
                </tr>
                <tr>
                    <td>Bananas</td>
                    <td>50</td>
                    <td>$1.50</td>
                </tr>
            </table>
            <h2>Subsection 1.1</h2>
            <p>Additional text in subsection 1.1 that is separated from the table and list.</p>
            <p>Here is a detailed list:</p>
            <ul>
                <li>Item 1: Description of item 1, which is quite detailed and important.</li>
                <li>Item 2: Description of item 2, which also contains significant information.</li>
                <li>Item 3: Description of item 3, another item that we don't want to split across chunks.</li>
            </ul>
        </div>
    </body>
</html>
"""

headers_to_split_on = [("h1", "Header 1"), ("h2", "Header 2")]

splitter = HTMLSemanticPreservingSplitter(
    headers_to_split_on=headers_to_split_on,
    max_chunk_size=50,
    elements_to_preserve=["table", "ul"],
)

documents = splitter.split_text(html_string)
print(documents)
````

#### 说明在此示例中，`HTMLSemanticPreservingSplitter` 确保整个表格和无序列表 (`<ul>`) 被完整保留在各自的文本块中。尽管块大小设置为50个字符，但该分割器能识别这些元素不应被拆分，从而保持其完整性。
在处理数据表或列表时，这一点尤为重要，因为拆分内容可能导致上下文丢失或混淆。生成的`Document`对象会保留这些元素的完整结构，从而确保信息的上下文关联性得以维持。

### 使用自定义处理程序`HTMLSemanticPreservingSplitter` 允许您为特定的 HTML 元素定义自定义处理程序。某些平台具有自定义的 HTML 标签，这些标签无法被 `BeautifulSoup` 原生解析。当出现这种情况时，您可以利用自定义处理程序轻松添加格式化逻辑。
这对于需要特殊处理的元素尤其有用，例如 `<iframe>` 标签或特定的 `data-` 元素。在本示例中，我们将为 `iframe` 标签创建一个自定义处理器，将其转换为类似 Markdown 的链接。

````python
def custom_iframe_extractor(iframe_tag):
    iframe_src = iframe_tag.get("src", "")
    return f"[iframe:{iframe_src}]({iframe_src})"


splitter = HTMLSemanticPreservingSplitter(
    headers_to_split_on=headers_to_split_on,
    max_chunk_size=50,
    separators=["\n\n", "\n", ". "],
    elements_to_preserve=["table", "ul", "ol"],
    custom_handlers={"iframe": custom_iframe_extractor},
)

html_string = """
<!DOCTYPE html>
<html>
    <body>
        <div>
            <h1>Section with Iframe</h1>
            <iframe src="https://example.com/embed"></iframe>
            <p>Some text after the iframe.</p>
            <ul>
                <li>Item 1: Description of item 1, which is quite detailed and important.</li>
                <li>Item 2: Description of item 2, which also contains significant information.</li>
                <li>Item 3: Description of item 3, another item that we don't want to split across chunks.</li>
            </ul>
        </div>
    </body>
</html>
"""

documents = splitter.split_text(html_string)
print(documents)
````

#### 说明在这个示例中，我们为`iframe`标签定义了一个自定义处理器，将其转换为类似Markdown的链接形式。当分割器处理HTML内容时，它会使用这个自定义处理器来转换`iframe`标签，同时保留表格和列表等其他元素。最终的`Document`对象展示了如何根据你提供的自定义逻辑处理iframe。
**重要提示**：在保留链接等元素时，需注意分隔符中不应包含英文句号`.`，或留空分隔符。因为`RecursiveCharacterTextSplitter`会按句号进行分割，这将导致链接被截断。请确保提供以`. `（句号加空格）作为分隔符的列表。
